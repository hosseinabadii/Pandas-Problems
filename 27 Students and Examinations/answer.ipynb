```python
import pandas as pd

students = pd.read_csv("students.csv")
subjects = pd.read_csv("subjects.csv")
examinations = pd.read_csv("examinations.csv")
```

```python
stu_sub_combination = pd.merge(students, subjects, how="cross")
stu_sub_combination.sort_values(by=["student_id", "subject_name"], inplace=True)
stu_sub_combination
```

```python
attended_exams = examinations.value_counts().reset_index(name="attended_exams")
attended_exams
```

```python
result = pd.merge(stu_sub_combination, attended_exams, how="left").fillna(0)
result["attended_exams"] = result["attended_exams"].astype("int")
result.drop(columns="subject_id")
```
