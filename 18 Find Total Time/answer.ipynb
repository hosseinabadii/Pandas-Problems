```python
import pandas as pd
employees = pd.read_csv("employees.csv")
employees
```

```python
df = employees.groupby(["emp_id", "event_day"]).agg("sum").reset_index()
df
```

```python
df["total_time"] = df["out_time"] - df["in_time"]
df
```

```python
df.sort_values(by="event_day", inplace=True)
df.rename(columns={"event_day": "day"}, inplace=True)
df[["day", "emp_id", "total_time"]]
```
